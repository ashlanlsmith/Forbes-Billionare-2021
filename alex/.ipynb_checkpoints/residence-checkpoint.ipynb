```python
# imports & config
import pandas as pd
import numpy as np
import geopandas as geo
import geoplot as gplot
import country_converter as converter
import mapclassify as mc
import matplotlib as mpl
import matplotlib.pyplot as plt
%config InlineBackend.figure_format ='retina'
```

```python
# load data
b = pd.read_csv('../data/billionaire.csv')
```

```python
# filter out families
family_filtered = b[~b['Name'].str.contains('family')]
```

```python
# determine billionaires per country
bpc = pd.DataFrame()
result = family_filtered['Country'].value_counts()
bpc['Country'] = result.index
bpc['Billionaire_count'] = result.values
cc = converter.CountryConverter()
bpc['ISO_A3'] = cc.convert(bpc['Country'],to='ISO3')
bpc.drop('Country', axis=1, inplace=True)
```

```python
# load shapefile
fp = 'countries.geojson'
country_geo = geo.read_file(fp)
```

```python
# merge geodataframe & dataframe
country_geo = pd.merge(country_geo, bpc, left_on='ISO_A3', right_on='ISO_A3', how='outer')
```

```python
country_geo['Billionaire_count'] = country_geo['Billionaire_count'].apply(lambda x: 0 if np.isnan(x) else x)
```

```python
fig, ax = plt.subplots(1, 1, figsize=(30, 30))
ud = mc.UserDefined(country_geo['Billionaire_count'], 
                    [10, 20, 30, 40, 50, 100, 200, 300, 400, 500])

gplot.choropleth(country_geo,
                hue='Billionaire_count',
                linewidth=1,
                scheme=ud,
                cmap='rainbow',
                legend=True,
                edgecolor='black',
                ax=ax)
ax.set_title('Billionaires per country', fontsize=22)
```
